```python
import wikipedia
from openai import OpenAI
client = OpenAI(api_key="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")

sections_contents = {
    "Collaborative and social functionality": wikipedia.page("Software as a Service").section("Collaborative and social functionality"),
    "Software as a Service": wikipedia.summary("Software as a Service"),
    "History": wikipedia.page("Software as a Service").section("History"),
    "Distribution and pricing": wikipedia.page("Software as a Service").section("Distribution and pricing"),
    "Architecture": wikipedia.page("Software as a Service").section("Architecture"),
    "Vertical vs horizontal SaaS": wikipedia.page("Software as a Service").section("Vertical vs horizontal SaaS"),
    "Characteristics": wikipedia.page("Software as a Service").section("Characteristics"),
    "Accelerated feature delivery": wikipedia.page("Software as a Service").section("Accelerated feature delivery"),
    "Open integration protocols": wikipedia.page("Software as a Service").section("Open integration protocols"),
    "OpenSaaS": wikipedia.page("Software as a Service").section("OpenSaaS"),
    "Adoption drivers": wikipedia.page("Software as a Service").section("Adoption drivers"),
    "Data escrow": wikipedia.page("Software as a Service").section("Data escrow"),
    "Criticism": wikipedia.page("Software as a Service").section("Criticism"),
    "See also": wikipedia.page("Software as a Service").section("See also"),
    "References": wikipedia.page("Software as a Service").references,
}

user_input = input("Enter the section title you want to retrieve(Titles are case sensitive): ")
if user_input in sections_contents:
    output = sections_contents[user_input]
else:
    print("Invalid section name. Please enter a valid section.")
    exit()

def summarize_content(content):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"Summarize into bullets of the following text:\n{content}"}
        ],
        temperature=0,
        max_tokens=300,
    )
    return response.choices[0].message.content

derived_output = summarize_content(output)
print("\nPrinting the summarized output:")
print(derived_output)

def paraphrase_content(content):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": f"parafrase the following text:\n{content}"}
        ],
        temperature=0,
        max_tokens=300,
    )
    return response.choices[0].message.content

derived_output = paraphrase_content(output)
print("\nPrinting the paraphased output:")
print(derived_output)

```
